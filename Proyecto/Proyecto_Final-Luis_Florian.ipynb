## **Predicción de precios de casas**

El siguiente proyecto consiste en crear modelos predictivos de regresión lineal uni-variable sencillos de la forma:

**$ y = f(x) =  mx+b $**

Donde:

$y$ = la variable dependiente\
$x$ = variable independiente\
$m$ = pendiente de la recta(parámetro del modelo)\
$b$ = intercepto(parámetro del modelo)

Se trabajara con un set de datos de muchas variables y se realiza un análisis exploratorio para visualizar y analizar los datos y entender cómo se comportan y luego  elegir las variables independientes “x” a trabajar(según el potencial predictivo de estas a través de medir la correlación), esto significa que aunque el dataset posee múltiples variables, en vez de crear un modelo multi-variable crearemos múltiples modelos uni-variable.

### **Modelo de Regresión Lineal de una variable**

```python
def training (x,y,epochs,imprimir_error_cada,lr):
    m,b = 0.0,0.0 # Inicializamos m y b con 0 para la primera iteración.
    parametros = np.array([m,b])  # [m,b]
    errors = np.zeros(epochs) # Error vector with epoch size.
    modelo = {}  # Estructura para guardar las predicciones de cada iteración   
    xx = x.reshape(len(x),1)
    xx = np.insert(xx,1,np.ones_like(x),axis=1) #Crear Matriz x de 2 columnas con "1"    
    
    for i in range(epochs):
        yhat = np.dot(xx,parametros) #yhat = mx+b         
        errors[i] = (np.average((y-yhat)**2))/2   
        gradientes = np.dot(yhat-y,xx)/len(y)
        parametros -= (lr * gradientes)
        modelo[i] = parametros.copy()
        
        if ((i+1) % imprimir_error_cada) == 0:
            print(f"Error Iteración {i+1}: ",errors[i])       
        
    return (modelo,errors)
```

### **Función para graficar errores**

```python
def graphErrores (errores):
    x = range(0,len(errores))
    y = errores[x]
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    
    ax.grid()
    ax.plot(x,errores, label = "Train Loss")
    
    ax.legend()
    ax.set_title("Losses")
    ax.set_ylabel('Errors')
    ax.set_xlabel('Iteracion')
    
    plt.show()    
    
```

### **Función de evolución del modelo entrenado en el tiempo**
Crear una función que nos permita visualizar con matplotlib cómo evoluciona el modelo entrenado en el tiempo


```python
def graphTraining(x,y,parametros, n):    
    
    #Ceiling division -(n // -d)
    filas_plots = -((len(parametros)//n) // -2)    
    
    print(f"filas:{type(filas_plots)}")
    
    f,axes = plt.subplots(filas_plots,2,sharex=False, figsize=(15,20))    
    
    xx = x.reshape(len(x),1)
    xx = np.insert(xx,1,np.ones_like(x),axis=1) #Crear Matriz x de 2 columnas con "1"     
    pos = 1 #Graph Position
    
    for i in range(n-1,len(modelo),n):  #Remembering first iteration starts at position 0 
        yhat = np.dot(xx,parametros[i]) #yhat = mx+b
        m = round(parametros[i][0],2)
        b = round(parametros[i][1],2)
        
        plt.subplot(filas_plots,2,pos)
        pos+=1
        
        
        plt.scatter(x,y)
        plt.plot(x,yhat, color='r')
        
        plt.title(f"Iteración No={i+1}, m={m}, b={b}")
    
    plt.show()
        
```

```python
epochs = 1000
error = 100
lr = 0.0001
#train[:,0] --> PriceSale
#train[:,1] -->OverallQual	
modelo, errores = training(train[:,1],train[:,0],epochs,error,lr)
graphErrores(errores)
graphTraining(train[:,1],train[:,0],modelo,170)
```
